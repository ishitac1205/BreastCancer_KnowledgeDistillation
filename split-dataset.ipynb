```python
import os
import glob
import random
from pathlib import Path
import shutil
from collections import defaultdict


random.seed(42)

img_paths = glob.glob('/kaggle/input/breakhis/BreaKHis_v1/BreaKHis_v1/histology_slides/breast/**/*.png', recursive=True)

subtype_paths = defaultdict(list)
for img_path in img_paths:
    img_name = Path(img_path).name
    subtype = img_name.split('-')[0]  
    subtype_paths[subtype].append(img_path)

val_size = 50    
test_size = 100  
print(f"Fixed sizes: Val={val_size}, Test={test_size}")
print("Train sizes will vary based on available images per class")

for subtype, paths in subtype_paths.items():
    if len(paths) < val_size + test_size:
        print(f"Warning: Class {subtype} has only {len(paths)} images (needs at least {val_size + test_size})")


base_dir = "BreakHis_splits"
os.makedirs(base_dir, exist_ok=True)
train_dir = os.path.join(base_dir, "train")
val_dir = os.path.join(base_dir, "val")
test_dir = os.path.join(base_dir, "test")

for split_dir in [train_dir, val_dir, test_dir]:
    os.makedirs(split_dir, exist_ok=True)

for subtype, paths in subtype_paths.items():
    random.shuffle(paths)  
    
    
    test_paths = paths[:test_size]
    val_paths = paths[test_size:test_size + val_size]
    train_paths = paths[test_size + val_size:]  # All remaining images go to train
    
   
    for split, split_paths in zip(["train", "val", "test"], [train_paths, val_paths, test_paths]):
        dest_dir = os.path.join(base_dir, split, subtype)
        os.makedirs(dest_dir, exist_ok=True)
        for path in split_paths:
            shutil.copy(path, dest_dir)

print("\nSplit counts:")
for split in ["train", "val", "test"]:
    split_dir = os.path.join(base_dir, split)
    print(f"\n{split.capitalize()} Set:")
    for subtype in subtype_paths.keys():
        subtype_dir = os.path.join(split_dir, subtype)
        if os.path.exists(subtype_dir):
            print(f"  {subtype}: {len(os.listdir(subtype_dir))} images")
```
